```python
import os
import cv2
import numpy as np
import pandas as pd

def get_img_list(folder):
    folder = folder + '/'
    class_list = os.listdir(folder)
    class_list.remove('.DS_Store')
    class_dict = dict(zip(class_list, list(range(0,len(class_list)))))
    image_list = []
    for cls in class_list:
        images = os.listdir(folder + cls)
        for i in images:
            img = cv2.imread(folder + cls + '/' + i)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img = cv2.resize(img,(128,128))
            image_list.append(img)
    image_array = np.array(image_list)
    image_array = image_array.astype('float32')
    image_array /= 255
    return class_list, class_dict, image_array
    
class_list, class_dict, image_array = get_img_list('data')
```

```python
def get_labels():
    labels = np.ones((image_array.shape[0]), dtype='int')
    count=0
    for cls in class_list:
        images = os.listdir('data/' + cls)
        labels[count:(len(images)+count)] = class_dict[cls]
        count += len(images)
#         print(len(images))
#         print(count)
    return labels

labels = get_labels()
```

# Pre-Processing

```python
from keras.utils import np_utils



def one_hot(labels, class_list):
    

    y = np_utils.to_categorical(labels, len(class_list))

    length_counts = []

    for i in y:
        if len(i) not in length_counts:
            length_counts.append(len(i))
    print('This should be 1: ', len(length_counts))
    return y

y = one_hot(labels, class_list)
```

```python
from sklearn.model_selection import train_test_split

def split(labels, class_list, image_array, random_state, test_size):
    y = one_hot(labels, class_list)
    X_train, X_test, y_train, y_test = train_test_split(
        image_array, y, test_size=test_size, random_state=random_state, shuffle=True)
    return X_train, X_test, y_train, y_test    
```

```python
X_train, X_test, y_train, y_test = split(labels, class_list, image_array, 42, .15)

print("X_train shape = {}".format(X_train.shape))
print("X_test shape = {}".format(X_test.shape))
```
