```python
def c_to_f(c):
    return c*(9/5)+32

countries_df['AverageTemperature_F'] = c_to_f(countries_df['AverageTemperature'])
countries_df['AverageTemperatureUncertainty_F'] = c_to_f(countries_df['AverageTemperatureUncertainty'])

countries_df.info()
```

```python
# Get list of incomplete years

monthdict={1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
baseline = set(range(1,13))
dictionary = {}

for i in range(countries_df['Year'].min(),countries_df['Year'].max()):
    df = countries_df[countries_df['Year'] == i]
    if len(df.groupby('Country').mean()['Month'].unique()) != 1:
        l = list(df.groupby('Country').mean()['Month'][df.groupby('Country').mean()['Month'] != 6.5].index)
        l2 = []
        for c in l:
            minidict = {}
            if baseline != set(list(df[df['Country'] == c]['Month'])):
                missing = list(baseline.difference(set(list(df[df['Country'] == c]['Month']))))
                minidict[c] = missing
                l2.append(minidict)
        dictionary[i] = l2
```
