```python
import pandas as pd
import numpy as np
import datetime


import seaborn as sns
import matplotlib.pyplot as plt

import pandas.tseries
```

```python
df = pd.read_pickle('cleaned_df.pkl')
```

```python
# Get list of incomplete years

monthdict={1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
baseline = set(range(1,13))
dictionary = {}

for i in range(df['Year'].min(),df['Year'].max()):
    df1 = df[df['Year'] == i]
    if len(df1.groupby('Country').mean()['Month'].unique()) != 1:
        l = list(df1.groupby('Country').mean()['Month'][df1.groupby('Country').mean()['Month'] != 6.5].index)
        l2 = []
        for c in l:
            d = {}
            if baseline != set(list(df1[df1['Country'] == c]['Month'])):
                missing = list(baseline.difference(set(list(df1[df1['Country'] == c]['Month']))))
                d[c] = missing
                l2.append(d)
        dictionary[i] = l2
```
